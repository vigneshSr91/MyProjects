```python
# 1D plots are hard to read
plt.close()

iris_setosa = iris.loc[iris["species"] == "setosa"]
iris_virginica = iris.loc[iris["species"] == 'virginica']
iris_versicolor = iris.loc[iris["species"] == "versicolor"]

plt.plot(iris_setosa["petal_length"], np.zeros_like(iris_setosa["petal_length"]))
plt.plot(iris_virginica["petal_length"], np.zeros_like(iris_virginica["petal_length"]))
plt.plot(iris_versicolor["petal_length"], np.zeros_like(iris_versicolor["petal_length"]))

plt.show()

# In the below plot, you could see many overlapping points and not easy to vizualize the distribution of data points
```

```python
# To avoid the above issue, we could use count of data points on the Y axis to display a Histogram

plt.close()
sns.FacetGrid(iris, hue="species", size=5) \
    .map(sns.distplot, "petal_length") \
        .add_legend()

plt.show()
```

```python
# From the above PDF ( Probablity density functions ), it is clear that it is more like a Univariate function
# sepal_length > sepal_width >> petal_length >>> petal_width

# It is better to choose a variable with less overlapping regions
```

```python
# Cumulative Distribution Function (Vs) Probablity Density Function

# CDF shows the probability in terms of distribution of data points

counts, bin_edges = np.histogram(iris_setosa['petal_length'], bins=10, density=True)

pdf = counts / sum((counts))
print(pdf)
print(bin_edges)

# Compute CDF
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:],cdf)

plt.show()
```

```python
counts, bin_edges = np.histogram(iris_setosa['petal_length'], bins=10, density=True)

pdf = counts / sum((counts))
print(pdf)
print(bin_edges)

# Compute CDF
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:],cdf)


counts, bin_edges = np.histogram(iris_virginica['petal_length'], bins=10, density=True)

pdf = counts / sum((counts))
print(pdf)
print(bin_edges)

# Compute CDF
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:],cdf)


counts, bin_edges = np.histogram(iris_versicolor['petal_length'], bins=10, density=True)

pdf = counts / sum((counts))
print(pdf)
print(bin_edges)

# Compute CDF
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:],cdf)


plt.show()
```

```python
# From the above plot, we can draw the below conclusions:

# if petal_length <=2, then Setosa
# if petal_length > 2 and petal_length <= 5, then Versicolor ( with 95% accuracy and 5% probability for error)
# if petal_length > 5 then Virginica ( with 90% accuracy and 10% probability for error)

# Hence CDF is more useful to gain more insights in comparison with PDF
```

```python
# Mean, Variance and Standard Deviations

print("Means:")
print("\n")
print("Mean of Setosa: ", np.mean(iris_setosa["petal_length"]))
print("Mean of Versicolor: ", np.mean(iris_versicolor["petal_length"]))
print("Mean of Virginica: ", np.mean(iris_virginica["petal_length"]))

# Calculate the Mean with an arbitrarily high outlier and see the effects on the Mean
print("Mean of Setosa after appending an outlier: ", np.mean(np.append(iris_setosa["petal_length"], 50)))

# Standard Deviations
print("\n")
print("Standard Deviations: ")
print("\n")
print("Standard Deviation of Setosa", np.std(iris_setosa["petal_length"]))
print("Standard Deviation of Versicolor", np.std(iris_versicolor["petal_length"]))
print("Standard Deviation of Virginica", np.std(iris_virginica["petal_length"]))
```

```python
# From the above we have seen that even one outlier in a small Dataset can cause a huge difference in the Mean, Variance and Standard Deviation Calculations
```

```python
print("Median:")
print("\n")
print("Median of Setosa: ", np.median(iris_setosa["petal_length"]))
print("Median of Versicolor: ", np.median(iris_versicolor["petal_length"]))
print("Median of Virginica: ", np.median(iris_virginica["petal_length"]))

# Calculate the Mean with an arbitrarily high outlier and see the effects on the Mean
print("Median of Setosa after appending an outlier: ", np.median(np.append(iris_setosa["petal_length"], 50)))
```

```python
# Median is not affected much by the outlier
"""
How Median is calculated ?
- First sort 
- Pick the middle value
-- If N is even, middle point and avg of left and right element
-- If N is odd, Pick the middle element
"""
```

```python
# Quantiles
print("Quantiles:")
print("\n")
print(np.percentile(iris_setosa["petal_length"], np.arange(0, 100, 25)))
print(np.percentile(iris_versicolor["petal_length"], np.arange(0, 100, 25)))
print(np.percentile(iris_virginica["petal_length"], np.arange(0, 100, 25)))
```

```python
# 90th Percentiles
print("90th Percentiles:")
print("\n")
print(np.percentile(iris_setosa["petal_length"], 90))
print(np.percentile(iris_versicolor["petal_length"], 90))
print(np.percentile(iris_virginica["petal_length"], 90))
```

```python
# Median Absolute Deviation
"""
- Take the median
- Calculate the square distance of each data point from the median 
- then Calculate the deviation( similar like standard deviation but not affected by outliers)
"""
from statsmodels import robust
print("\nMedian Absolute Deviation")
print("\n")
print(robust.mad(iris_setosa["petal_length"]))
print(robust.mad(iris_versicolor["petal_length"]))
print(robust.mad(iris_virginica["petal_length"]))
```
